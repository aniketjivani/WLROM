Generate pilot data at a single parameter setting, and build a CNN to fit the 2D convection-reaction-diffusion data (all timesteps used in training, we are not currently trying to extrapolate anything).

This is more experimental, Notebook 01 onwards will have more systematic experiments with models, architectures and so on.

```python
import numpy as np
import matplotlib.pyplot as plt
import toy_utils as ut
import os
```

```python
nxHF = 64
nyHF = 64

uHFFinal, vHFFinal = ut.getVelocitiesGeneric(nxHF, nyHF)

input_list_HF_pilot = np.loadtxt("./input_list_2D_example.txt")

NHF = input_list_HF_pilot.shape[0]
```

```python
tf = 3.0
```

```python
alpha = 1e-2
```

```python
x, y, xm, ym, XM, YM, dx, dy, dxi, dyi, dxi2, dyi2 = ut.getGridQuantities(nxHF, nyHF)
```

```python
theta_s = 0.01
theta_h = 0.05

theta_x1 = 0.3
theta_y1 = 0.55

theta_x2 = 0.7
theta_y2 = 0.4

theta_x3 = 0.5
theta_y3 = 0.6
```

```python
CFL = 0.8

maxU = np.max(np.abs(uHFFinal))
maxU = np.max([maxU, np.max(np.abs(vHFFinal))])
dt_c = CFL * dx / maxU  # dt based on convective CFL
dt_v = CFL * dx ** 2 / alpha / 4  # dt based on viscous CFL


dt = min(dt_c, dt_v)   # Simulation timestep
dt
```

```python
NHF # not used currently.
```

```python
times = np.arange(0, tf + dt, dt)
np.arange(0, tf + dt, dt).shape # these are the timesteps at which solution is available.
```
